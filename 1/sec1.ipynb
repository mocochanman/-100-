# データの読み込み、結合

```python
import pandas as pd
customer_master = pd.read_csv('customer_master.csv')
customer_master.head()
```

```python
item_master = pd.read_csv('item_master.csv')
transaction_1 = pd.read_csv('transaction_1.csv')
transaction_detail_1 = pd.read_csv('transaction_detail_1.csv')
```

```python
#データのユニオン（結合なので縦かな）
transaction_2 = pd.read_csv('transaction_2.csv')
transaction_detail_2 = pd.read_csv('transaction_detail_2.csv')

transaction = pd.concat([transaction_1, transaction_2], ignore_index=True)
transaction_detail = pd.concat([transaction_detail_1, transaction_detail_2], ignore_index=True)
#ignore indexは連結方向のラベルの振り直し
```

```python
#データのジョイン（横方向の結合）
join_data = pd.merge(transaction_detail, transaction[["transaction_id", "payment_date", "customer_id"]], on="transaction_id", how="left")
join_data.head()
```

```python
join_data = pd.merge(join_data, customer_master, on="customer_id", how="left")
join_data = pd.merge(join_data, item_master, on="item_id", how="left")
join_data.head()
```

## ポイントは必要な属性は何なのか、重複しているデータはないか

# 必要なデータ列の作成（売上列の作成）

```python
join_data["price"] = join_data["quantity"] * join_data["item_price"]
join_data[["price", "quantity", "item_price"]].head()
```

# データ検算

```python
print(join_data["price"].sum())
print(transaction["price"].sum())

join_data["price"].sum() == transaction["price"].sum()
```

# 月別集計

```python
#データ型の確認
join_data.dtypes
```

### object → datetimeに変更し、年月列の作成をして月毎に集計する

```python
join_data["payment_date"] = pd.to_datetime(join_data["payment_date"])
join_data["payment_month"] = join_data["payment_date"].dt.strftime("%Y%m")
join_data[["payment_date", "payment_month"]]
```

```python
#月毎に集計する
join_data.groupby("payment_month").sum()["price"]

import matplotlib.pyplot as plt
import numpy as np
uriage = np.array(list(join_data.groupby("payment_month").sum()["price"]))
plt.plot(uriage)
plt.show()
```

### 月別かつ商品別の集計

```python
join_data.groupby(["payment_month", "item_name"]).sum()["price"]
#見にくいからテーブル化する
pd.pivot_table(join_data, index="item_name", columns="payment_month", values=["price", "quantity"], aggfunc="sum")
```

# 商品別の売上推移の可視化

```python
graph_data_1 = pd.pivot_table(join_data, index="payment_month", columns="item_name", values="price", aggfunc="sum")
graph_data_1
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
plt.title("Monthly sales data of products")
plt.xlabel("month")
plt.ylabel("sales")
plt.plot((graph_data_1.index), graph_data_1["PC-A"], label="PC-A")
plt.plot((graph_data_1.index), graph_data_1["PC-B"], label="PC-B")
plt.plot((graph_data_1.index), graph_data_1["PC-C"], label="PC-C")
plt.plot((graph_data_1.index), graph_data_1["PC-D"], label="PC-D")
plt.plot((graph_data_1.index), graph_data_1["PC-E"], label="PC-E")
plt.legend()
plt.show()
```
